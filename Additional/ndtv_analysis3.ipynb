```python
def touch(x):
    if (x['Form factor']=='Touchscreen') & (x['Touchscreen']!=x['Touchscreen']):
        return 'Yes'
    else:
        return x['Touchscreen']
```

```python
data['Touchscreen']=data[['Form factor','Touchscreen']].apply(touch,axis=1)
```

```python
def four(x):
    if (x['3G']!=x['3G']) & (x['4G/ LTE']!=x['4G/ LTE']):
        return 'No'
    else:
        return x['4G/ LTE']
```

```python
def three(x):
    if (x['3G']!=x['3G']):
        return 'No'
    else:
        return x['3G']
```

```python
data['4G/ LTE']=data[['3G','4G/ LTE']].apply(four,axis=1)
```

```python
import re
def matcher_blue(x):
    if(x['Bluetooth']!=x['Bluetooth']):
        return x['Bluetooth']
    if(re.match('No*.',x['Bluetooth'])):
        return 'No'
    return 'Yes'  
```

```python
data['Bluetooth'] = data.apply(matcher_blue, axis=1)
```

```python
data['Bluetooth'].value_counts()
```

```python
data[(data['Bluetooth']!=data['Bluetooth']) & (data['Wi-Fi']!=data['Wi-Fi'])]
```

```python
def blue(x):
    if (x['Bluetooth']!=x['Bluetooth']) & (x['Wi-Fi']!=x['Wi-Fi']):
        return 'No'
    else:
        return x['Bluetooth']
```

```python
def wifi(x):
    if (x['Wi-Fi']!=x['Wi-Fi']):
        return 'No'
    else:
        return x['Wi-Fi']
```

```python
data['Bluetooth']=data[['Bluetooth','Wi-Fi']].apply(blue,axis=1)
```

```python
data['Wi-Fi']=data.apply(wifi,axis=1)
```

```python
def lblue(x):
    if (x['Wi-Fi']=='Yes') & (x['Bluetooth']!=x['Bluetooth']):
        return 'Yes'
    else:
        return x['Bluetooth']
```
